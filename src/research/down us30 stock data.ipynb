```python
client = Client(API_KEY, API_SECRET)

def get_data(symbol):
    data = []

    while True:
        if start_time >= today:
            break

        res = client.get_klines(
            symbol=symbol, 
            start_time=start_time,
            interval=CandlesticksChartInervals.HOUR, limit=1000)
        
        if len(res) == 0:
            break
        
        data.extend(res)
        start_time = datetime.fromtimestamp(res[-1][0] / 1000) + timedelta(hours=1)

    df = pd.DataFrame(data, columns=['time', 'open', 'high', 'low', 'close', 'volume'])
    df['time'] = pd.to_datetime(df['time'], unit='ms')
    df[['open', 'high', 'low', 'close', 'volume']] = df[['open', 'high', 'low', 'close', 'volume']].astype('float32')
    df.set_index('time', inplace=True)

    return df
```
